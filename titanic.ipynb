```python
#Importing the required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Reading the Dataset
df = pd.read_csv("/train.csv")
```

```python
print("Displaying the first 5 rows of the dataset: ")
df.head()
```

```python
print("\nSummary Statistics: ")
df.describe()
```

```python
print("Information about the dataset: \n\n")
df.info()
```

# **Cleaning the Dataset**

```python
print("\nDetecting missing values: ")
df.isnull().sum()
```

```python
#Numerical column
df.shape
```

```python
df['Age']= df['Age'].fillna(df['Age'].median())
```

```python
df['Age'].isnull().sum()
```

```python
df= df.drop(['Cabin'],axis=1)
```

```python
df
```

```python
df['Embarked']= df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df['Embarked'].isnull().sum()
```

```python
df
```

# Exploring the Dataset

```python
plt.figure(figsize=(9, 4))
sns.histplot(df["Age"], kde=True, color='#9400D3')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(9, 4))
sns.countplot(data=df, x="Sex", hue="Sex",palette={"male": "Yellow", "female": "purple"})
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Distribution",labels=["Male", "Female"])
plt.show()
```

```python
df['Age_Binned'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
bins = [0, 15, 30, 45, 60, 75, float('inf')]
labels = ['0-15', '16-30', '31-45', '46-60', '61-75', '76+']

plt.figure(figsize=(9, 4))
sns.scatterplot(data=df, x="Age", y="Fare", hue="Age_Binned", palette=["blue", "green", "yellow", "orange", "red", "purple"])
plt.title("Scatter_plot")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Age Group",)
plt.show()
```

```python
# Explore the relationship between 'Survived' and other categorical features
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Survived', hue='Sex')
plt.title('Survival Count by Gender')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Survived', hue='Pclass')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Survived', hue='Embarked')
plt.title('Survival Count by Embarkation Port')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
# Explore the distribution of 'Fare'
plt.figure(figsize=(9, 4))
sns.histplot(df["Fare"], kde=True, color='green')
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
# Explore the relationship between 'Fare' and 'Survived'
plt.figure(figsize=(9, 4))
sns.boxplot(data=df, x='Survived', y='Fare')
plt.title("Survival by Fare")
plt.xlabel("Survived")
plt.ylabel("Fare")
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python

numerical_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(8, 6))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

```python

plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='SibSp', hue='Survived')
plt.title('Survival Count by Number of Siblings/Spouses')
plt.xlabel('Number of Siblings/Spouses')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Parch', hue='Survived')
plt.title('Survival Count by Number of Parents/Children')
plt.xlabel('Number of Parents/Children')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```
